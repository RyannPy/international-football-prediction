## 2. Definisi Setiap Data

Kita perlu memahami maksud dan definisi dari setiap data dan kolom.

```python
results.head()
```

### Insight
Tabel `results`

Tabel ini menyimpan data historis hasil pertandingan internasional. Data ini terdiri dari 9 kolom yang merangkum rincian setiap laga:

*   **`date`**: Tanggal kapan pertandingan tersebut dilaksanakan (dengan format Tahun-Bulan-Tanggal).
*   **`home_team`**: Nama negara atau tim yang bertindak sebagai tuan rumah (bermain di kandang).
*   **`away_team`**: Nama negara atau tim yang bertindak sebagai tamu (lawan).
*   **`home_score`**: Skor atau jumlah gol yang berhasil dicetak oleh tim tuan rumah.
*   **`away_score`**: Skor atau jumlah gol yang berhasil dicetak oleh tim tamu.
*   **`tournament`**: Jenis kompetisi atau ajang pertandingan tersebut digelar (misalnya *Friendly* yang berarti laga persahabatan).
*   **`city`**: Nama kota tempat pertandingan diselenggarakan.
*   **`country`**: Nama negara tempat pertandingan diselenggarakan.
*   **`neutral`**: Keterangan (*True/False*) yang menunjukkan apakah pertandingan dimainkan di tempat netral. Jika bernilai `False`, berarti pertandingan benar-benar dilakukan di kandang tim tuan rumah, bukan di lokasi pihak ketiga.

```python
former_names.head()
```

### Insight
Tabel `former_names`

Tabel ini berfungsi sebagai kamus sejarah yang melacak perubahan nama suatu negara atau tim nasional dari waktu ke waktu. Tabel ini sangat berguna untuk memastikan data pertandingan masa lalu (dengan nama lama) tetap bisa dihubungkan dengan negara modernnya. Data ini terdiri dari 4 kolom:

*   **`current`**: Nama resmi negara atau tim tersebut saat ini.
*   **`former`**: Nama lama, julukan historis, atau entitas politik terdahulu dari negara tersebut.
*   **`start_date`**: Tanggal kapan nama lama tersebut mulai digunakan atau tercatat secara resmi.
*   **`end_date`**: Tanggal kapan penggunaan nama lama tersebut berakhir (kapan entitas tersebut berganti nama atau bubar).



```python
goalscorers.head()
```

### Insight
Tabel `goalscorers`

Tabel ini menyimpan catatan detail dari setiap gol yang tercipta di berbagai pertandingan. Ini adalah data tingkat pemain (bukan sekadar hasil akhir tim). Data ini terdiri dari 8 kolom:

*   **`date`**: Tanggal kapan pertandingan tersebut berlangsung.
*   **`home_team`**: Nama tim tuan rumah.
*   **`away_team`**: Nama tim tamu.
*   **`team`**: Nama tim dari pemain yang mencetak gol (berguna untuk mencocokkan apakah gol dicetak oleh tim tuan rumah atau tamu).
*   **`scorer`**: Nama lengkap atau panggil dari pemain yang berhasil mencetak gol.
*   **`minute`**: Menit pertandingan ke berapa gol tersebut bersarang di gawang.
*   **`own_goal`**: Indikator (*True/False*) apakah gol tersebut merupakan gol bunuh diri. Jika `True`, berarti pemain secara tidak sengaja memasukkan bola ke gawangnya sendiri.
*   **`penalty`**: Indikator (*True/False*) apakah gol tersebut dicetak melalui titik putih (tendangan penalti).


```python
shootouts.head()
```

### Insight
Tabel `shootouts`

Tabel ini secara khusus merangkum hasil akhir dari pertandingan-pertandingan yang berakhir imbang dan harus ditentukan melalui drama adu penalti. Data ini terdiri dari 5 kolom:

*   **`date`**: Tanggal terjadinya pertandingan adu penalti tersebut.
*   **`home_team`**: Nama tim yang bertindak sebagai tuan rumah.
*   **`away_team`**: Nama tim lawan (tamu).
*   **`winner`**: Nama tim yang pada akhirnya memenangkan babak adu penalti.
*   **`first_shooter`**: Nama tim yang mendapatkan giliran pertama untuk menendang bola dalam adu penalti. (Catatan: Nilai `NaN` yang terlihat pada baris awal menunjukkan bahwa informasi ini tidak tercatat atau kosong pada pertandingan tersebut).
