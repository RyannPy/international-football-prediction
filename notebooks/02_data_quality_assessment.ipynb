## 4. Kesimpulan Akhir dan Rencana Eksekusi

Berdasarkan seluruh tahapan pengecekan **Informasi Umum Data** dan **Investigasi** mendalam yang telah dilakukan, kita kini memiliki pemahaman yang utuh mengenai kondisi awal dataset. Secara keseluruhan, dataset ini berkualitas baik, namun memiliki beberapa karakteristik unik, anomali historis, dan data masa depan yang wajar ditemui pada *real-world data*.

Tujuan utama kita adalah menyiapkan data untuk pemodelan prediksi hasil pertandingan. Oleh karena itu, berikut adalah keputusan final atau *Action Plan* yang akan dieksekusi pada proses *Data Cleaning* sesungguhnya:

**A. Standardisasi Format**
* Mengonversi seluruh kolom yang memuat informasi waktu (`date` pada tabel `results`, `goalscorers`, dan `shootouts`; serta `start_date` & `end_date` pada tabel `former_names`) dari format *string* menjadi tipe data *datetime*.

**B. Penanganan Data Kosong (*Missing Values*)**
* **Tabel `results`**: Menghapus (*drop*) 44 baris pertandingan FIFA World Cup 2026. Data ini adalah jadwal masa depan yang skornya belum ada, sehingga secara absolut tidak dapat digunakan sebagai target pelatihan model.
* **Tabel `goalscorers`**: Membiarkan data kosong pada kolom `scorer` dan `minute`. Ketidaklengkapan ini murni karena rekam jejak historis pada turnamen lama (seperti Oceania Nations Cup 1980) dan tidak memengaruhi dataset utama yang akan digunakan untuk pemodelan prediksi.
* **Tabel `shootouts`**: Menghapus kolom `first_shooter` secara keseluruhan (*drop column*). Selain karena persentase data kosongnya terlalu mendominasi, fitur ini juga tidak relevan dengan ruang lingkup *modelling* kita.

**C. Penanganan Data Duplikat**
* **Tabel `goalscorers`**: Mempertahankan 82 baris yang terdeteksi sebagai duplikat. Berdasarkan investigasi, data ini merepresentasikan kejadian nyata di mana dua gol berbeda dicetak oleh pemain yang sama dalam waktu berdekatan pada masa *injury time* (misalnya 45+1' dan 45+2'), yang sayangnya sama-sama dibulatkan pencatatannya menjadi menit ke-45.

> **Catatan Penting:** Dokumentasi *insight* ini memastikan bahwa tidak ada data historis valid yang terbuang sia-sia akibat pembersihan buta (*blind dropping*). Langkah-langkah eksekusi di atas siap diimplementasikan pada tahap pembersihan data (tahap selanjutnya) agar dataset final valid, bersih, dan siap masuk ke fase *Exploratory Data Analysis* (EDA) serta pemodelan *Machine Learning*.
